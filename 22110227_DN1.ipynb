```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

df = pd.read_csv('/content/books.csv')


d=dict(df)

year={}

for i in df.loc[:,"original_publication_year"]:
  if i in year:
    year[i]+=1
  else:
    year[i]=1
print(year)
year = {k:v for k,v in year.items() if not math.isnan(v)}

# Calculate books per decade
books_per_decade = {}
for year, count in year.items():
    decade = int(year//10)*10
    if decade in books_per_decade:
        books_per_decade[decade] += count
    else:
        books_per_decade[decade] = count

print(books_per_decade)
```

```python
import pandas as pd

# Read in the CSV file
fil = pd.read_csv('/content/books.csv')
num_books_by_author = fil.groupby("authors")['book_id'].nunique().reset_index()
num_books_by_author = num_books_by_author.rename(columns={'book_id': 'num_books'})
num_books_by_author = num_books_by_author.sort_values(by='num_books', ascending=False)
avg_rating_by_author = fil.groupby("authors")['average_rating'].mean().reset_index()
avg_rating_by_author = avg_rating_by_author.rename(columns={'average_rating': 'avg_rating'})
avg_rating_by_author = avg_rating_by_author.sort_values(by='avg_rating', ascending=False)

authors_df = pd.merge(num_books_by_author, avg_rating_by_author, on='authors')
print(authors_df.head(20))
plt.figure().set_figwidth(15)
plt.bar(authors_df.head(10)['authors'],authors_df.head(10)['avg_rating'])

top_20_mean_rating = authors_df.head(20)['avg_rating'].mean()

all_authors_mean_rating = authors_df['avg_rating'].mean()
print("Average rating of top 20 authors:", top_20_mean_rating)
print("Average rating of all authors:", all_authors_mean_rating)
```
